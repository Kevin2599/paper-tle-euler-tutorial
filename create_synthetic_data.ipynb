# Generate synthetic magnetic data

This notebook generates some synthetic total field magnetic anomaly data. We use the forward modeling capabilities of [Fatiando a Terra](http://fatiando.org) to create models of polygonal prisms.

```python
import json
import cPickle as pickle
import numpy as np
from IPython.display import Image
from fatiando.mesher import PolygonalPrism
from fatiando.gravmag import polyprism
from fatiando  import gridder, utils
from fatiando.vis import myv, mpl
import fatiando
print("Using Fatiando a Terra version {0}".format((fatiando.version)))
```

```python
area = [5000, 25000, 5000, 25000]
bounds = [0, 30000, 0, 30000, 0, 5000]
```

```python
mpl.figure()
mpl.axis('scaled')
mpl.square(area)
dike_verts = mpl.draw_polygon(bounds[:4], mpl.gca(), xy2ne=True)
```

```python
dike = PolygonalPrism(dike_verts, 0, 5000, {'magnetization': 10})

myv.figure()
myv.polyprisms([dike])
myv.axes(myv.outline(bounds), ranges=[b*0.001 for b in bounds], nlabels=5)
myv.wall_north(bounds)
myv.wall_bottom(bounds)
myv.show()
```

```python
mpl.figure()
mpl.axis('scaled')
mpl.square(area)
mpl.polygon(dike, xy2ne=True)
sill_verts = mpl.draw_polygon(bounds[:4], mpl.gca(), xy2ne=True)
```

```python
sill = PolygonalPrism(sill_verts, 1000, 1500, {'magnetization': 10})

myv.figure()
myv.polyprisms([sill, dike])
myv.axes(myv.outline(bounds), ranges=[b*0.001 for b in bounds], nlabels=5)
myv.wall_north(bounds)
myv.wall_bottom(bounds)
myv.show()
```

```python
mpl.figure()
mpl.axis('scaled')
mpl.square(area)
mpl.polygon(dike, xy2ne=True)
mpl.polygon(sill, xy2ne=True)
batholith_verts = mpl.draw_polygon(bounds[:4], mpl.gca(), xy2ne=True)
```

```python
batholith = PolygonalPrism(batholith_verts, 500, 4000, {'magnetization': 2})

myv.figure()
myv.polyprisms([sill, dike, batholith], prop='magnetization')
myv.axes(myv.outline(bounds), ranges=[b*0.001 for b in bounds], nlabels=5)
myv.wall_north(bounds)
myv.wall_bottom(bounds)
myv.show()
```

```python
with open('dike.json', 'w') as f:
    json.dump(dike_verts.tolist(), f)
with open('sill.json', 'w') as f:
    json.dump(sill_verts.tolist(), f)
with open('batholith.json', 'w') as f:
    json.dump(batholith_verts.tolist(), f)
```

```python
shape = (100, 100)
x, y, z = gridder.regular(area, shape, z=-300)
inc, dec = -15, 30
tf = utils.contaminate(polyprism.tf(x, y, z, [dike, sill, batholith], inc, dec), 5)
```

```python
mpl.figure()
mpl.axis('scaled')
levels = range(-760, 760, 20)
mpl.contourf(y, x, tf, shape, 60, cmap=mpl.cm.RdBu_r)
cb = mpl.colorbar()
cb.set_label('nT')
mpl.m2km()
mpl.xlabel('East (km)')
mpl.ylabel('North (km)')
mpl.savefig('synthetic_data.png')
mpl.show()
Image(filename='synthetic_data.png')
```

```python
with open('metadata.json', 'w') as f:
    json.dump(dict(area=area, bounds=bounds, inc=inc, dec=dec, shape=shape), f)
np.savetxt('synthetic_data.txt', np.transpose([x, y, z, tf]))
with open('model.pickle', 'w') as f:
    pickle.dump([dike, sill, batholith], f)
```
